```python
### Tratamiento de datos
import glob
import os
import pandas as pd
import numpy as np
import calendar
from contextlib import contextmanager
from datetime import datetime
import math

###############################################################################################################################################
#### Produccion Plan
os.chdir('D:/BI Plannig/01 Imput/02 Plan Produccion')
P_P = []
csv_files = glob.glob('*.xlsx')

for filename in csv_files:
    data = pd.read_excel(filename,sheet_name='Input marito')#,dtype={'Sugerido D+2':np.float64,'Sugerido revisado D + 2':np.float64})
    P_P.append(data)
P_P = pd.concat(P_P,ignore_index=True)

P_P = P_P.drop(['Descripción'], axis=1)
P_P = P_P.dropna(subset=['Material '])
P_P = P_P.set_index(('Material '))

## Transformar en una entrada
sr = P_P.stack()
st = pd.DataFrame(sr).reset_index()
st['level_1'] = st.level_1.astype(str)
t = st.level_1.str.split(" ", n=1, expand=True)
st["Fecha"] = t[0]
st = st[st["Fecha"]!="sem"]
st.rename(columns = {0:'Valor','Material ':'Material'} ,inplace=True)
PP_ok = st[['Material','Fecha','Valor']]

### SKU
PP_ok.Material  = PP_ok.Material.astype(str)
Titulo = PP_ok[["Material"]]
Titulo = Titulo["Material"].str.split(".", n=1 , expand=True)
PP_ok['Material'] =  Titulo[0].astype(str)
PP_ok.Valor = PP_ok.Valor.astype(str)
PP_ok["Valor"] = PP_ok.Valor.str.replace(r"º","")

### Fecha
PP_ok.Fecha  = PP_ok.Fecha.astype(str)
Titulo = PP_ok[["Fecha"]]
Titulo = Titulo["Fecha"].str.split("-", n=2 , expand=True)
PP_ok['Fecha'] =  Titulo[0].astype(str) + Titulo[1].astype(str) + Titulo[2].astype(str)

PP_ok.Valor = PP_ok.Valor.astype(float)
PP_ok.Valor = PP_ok.Valor.round(3)
PP_ok.Valor = PP_ok.Valor*1000
PP_ok.Valor = PP_ok.Valor.round(0)
PP_ok = PP_ok[PP_ok["Valor"]>0]

PP_ok.to_csv("D:/BI Plannig/02 ETL/02 DATA/03 Produccion Plan/Produccion_Plan.txt")

#####################################################################################################################################################
```

```python
################################################################################################################################################

##### Maquila Plan
os.chdir('D:/BI Plannig/01 Imput/04 Plan de maquila')
M_P = []
csv_files = glob.glob('*.xlsx')

for filename in csv_files:
    data = pd.read_excel(filename,sheet_name='Input marito')#,dtype={'Sugerido D+2':np.float64,'Sugerido revisado D + 2':np.float64})
    M_P.append(data)
M_P = pd.concat(M_P,ignore_index=True)

M_P = M_P.drop(['SAP'], axis=1)
M_P = M_P.drop(['Producto'], axis=1)
M_P = M_P.dropna(subset=['Truck'])
M_P = M_P.set_index(('Truck'))

## Transformar en una entrada
sr = M_P.stack()
st = pd.DataFrame(sr).reset_index()
st.level_1 = st.level_1.astype(str)
t = st.level_1.str.split(" ", n=1, expand=True)
st["CP"] = t[0]
st = st[st["CP"]!="sem"]
st = st[st[0]>0]
st.rename(columns = {0:'Valor','Truck':'Material','CP':'Fecha'} ,inplace=True)
MP_ok = st[['Material','Fecha','Valor']]

MP_ok.Valor = MP_ok.Valor.astype(float)
MP_ok.Valor = MP_ok.Valor.round(3)
MP_ok.Valor = MP_ok.Valor*1000
MP_ok.Valor = MP_ok.Valor.round(0)

MP_ok.Material = MP_ok.Material.astype(str)

### Fecha
MP_ok.Fecha  = MP_ok.Fecha.astype(str)
Titulo = MP_ok[["Fecha"]]
Titulo = Titulo["Fecha"].str.split("-", n=2, expand=True)
MP_ok['Fecha'] =  Titulo[0].astype(str) + Titulo[1].astype(str) + Titulo[2].astype(str)

MP_ok.to_csv("D:/BI Plannig/02 ETL/02 DATA/04 Maquila Plan/Maquila_Plan.txt")
```
