```python
#### IMPORTACIONES REAL
os.chdir('D:/BI Plannig/01 Imput/01 Importaciones/Imp Real')
Imp = []
csv_files = glob.glob('*.xlsx')

for filename in csv_files:
    data = pd.read_excel(filename,sheet_name='Consolidado')#,dtype={'Sugerido D+2':np.float64,'Sugerido revisado D + 2':np.float64})
    #data['Fecha_car']=filename
    Imp.append(data)
Imp = pd.concat(Imp,ignore_index=True)

### Limpieza inicial
Imp =Imp[["SKU Truck", "Ingreso estimado", "Puerto de llegada real", "Puerto de Origen", "HLs por recibir"]]
Imp.rename(columns = {'SKU Truck': 'Material','Ingreso estimado': 'Fecha_t','Puerto de llegada real':'P_llegada','Puerto de Origen':'P_Origen', 'HLs por recibir':'Imp'} ,inplace=True)
Imp.Imp = Imp.Imp.fillna(0)
Imp = Imp.dropna(subset=['Imp'])
Imp = Imp.drop(Imp[Imp['Imp']=='-'].index)

### Valor
Imp.Imp  =Imp.Imp.astype(float)
Imp.Imp = Imp.Imp.round(3)
Imp.Imp = Imp.Imp*1000
Imp.Imp = Imp.Imp.round(0)

### Fecha
Imp.Fecha_t  = Imp.Fecha_t.astype(str)
Titulo = Imp[["Fecha_t"]]
Titulo = Titulo["Fecha_t"].str.split("-", n=2, expand=True)
Imp['Fecha_t'] =  Titulo[0].astype(str) + Titulo[1].astype(str) + Titulo[2].astype(str)
Titulo = Imp["Fecha_t"].str.split(" ", n=1, expand=True)
Imp['Fecha_t'] =  Titulo[0].astype(str)
Imp = Imp.fillna(0)
Imp = Imp[Imp["Fecha_t"]!="00:00:00NoneNone"]
Imp = Imp[Imp["Fecha_t"]!="nanNoneNone"]

### SKU
Imp.Material = Imp.Material.astype(str)
Titulo = Imp["Material"].str.split(".", n=1, expand=True)
Imp['Material'] =  Titulo[0].astype(str)

Imp_r = Imp
Imp_r.to_csv("D:/BI Plannig/02 ETL/02 DATA/06 Importaciones/Importaciones_Real.txt")
Imp_r
```
